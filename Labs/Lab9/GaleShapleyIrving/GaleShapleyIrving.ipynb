#Stable Roommates Algorithm

```
def first_phase(players):

    steps_taken = 0

    proposed_to = {k: None for k in players}

    while any(k in proposed_to and proposed_to[k] == None for k in players):

        for i in players:

            steps_taken += 1

            if not any(k in proposed_to and proposed_to[k] == i for k in players):

                proposer = i

                break

        if len(players[proposer]) == 0:

            return False, False

        next_choice = players[proposer][0]

        if proposed_to[next_choice] == None:

            proposed_to[next_choice] = proposer

        else:

            candidate = proposed_to[next_choice]

            if players[next_choice].index(proposer) < players[next_choice].index(candidate):

                proposed_to[next_choice] = proposer

                players[next_choice].remove(candidate)
                players[candidate].remove(next_choice)

            else:

                players[next_choice].remove(proposer)
                players[proposer].remove(next_choice)

    for proposer in players:

        steps_taken += 1

        next_choice = players[proposer][0]

        lower_prospects = players[next_choice][players[next_choice].index(proposer)+1:]
        players[next_choice] = players[next_choice][:players[next_choice].index(proposer)+1]
        
        for i in lower_prospects:

            players[i].remove(next_choice)

    return players, steps_taken

def second_phase(players, steps_taken):

    if players == False:
        return "No solution exists"

    while True:

        for proposer in players:

            steps_taken += 1

            p = []
            q = []

            if len(players[proposer]) > 1:

                p.append(proposer)
                q.append(players[proposer][1])
                i = 0

                while True:

                    p.append(players[q[i]][-1])
                    
                    if p.count(players[q[i]][-1]) == 2:

                        for i in range(len(q)-1,-1,-1):

                            if p[i+1] in players[q[i]]:

                                players[q[i]].remove(p[i+1])

                            if q[i] in players[p[i+1]]:

                                players[p[i+1]].remove(q[i])

                        break

                    if len(players[p[i+1]]) == 1:
                        break

                    q.append(players[p[i+1]][1])

                    i += 1

        if any(k in players and players[k] == [] for k in players):

            return False, False

        elif all(k in players and len(players[k]) == 1 for k in players):

            return players

def roomate_matching(players):
    players, steps_taken = first_phase(players)
    if players == False:
        return "No solution exists"
    players = second_phase(players, steps_taken)
    if players == False:
        return "No solution exists"
    return players, ("Steps Taken: ", steps_taken)
```
